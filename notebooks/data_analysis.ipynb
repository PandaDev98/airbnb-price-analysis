```python
import pandas as pd

df = pd.read_csv(r"C:\Users\juand\OneDrive\Desktop\Python Course\Just fun\Airbnb- Analyzer\airbnb-price-analysis\data\listings.csv")
print(df.head())
```

```python
df = df[["id", "name", "host_id", "host_name", "neighbourhood", "room_type", "price", "minimum_nights", "number_of_reviews"]]
df = df.dropna()
df["price"] = pd.to_numeric(df["price"], errors="coerce")

print(df.describe())
df.groupby("neighbourhood")["price"].mean().sort_values(ascending=False).head(10)

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 5))
sns.histplot(df["price"], bins=50, kde=True)
plt.xlim(0, 500)
plt.title("Distribution of Airbnb Prices")
plt.show()
```

```python
sns.countplot(y=df["room_type"], order=df["room_type"].value_counts().index)
plt.title("Room Type Distribution")
plt.show()
```
